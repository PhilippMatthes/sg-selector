```python
import json
import os

with open('../../../data/example_routes_osm.json') as f:
    routes_json = json.load(f)
    
routes_with_bindings = []
    
for file in os.listdir("../../../data/bindings/"):
    filename = os.fsdecode(file)
    if not filename.endswith(".json"): 
        continue
    route_id = filename.replace(".json", "")
    routes_with_bindings.append(routes_json[int(route_id)])
    
routes_geojson = {
    "type": "FeatureCollection",
    "features": []
}
    
for route in routes_with_bindings:
    routes_geojson["features"].append({
        "type": "Feature",
        "properties": {},
        "geometry": {
            "type": "LineString",
            "coordinates": [[point["lon"], point["lat"]] for point in route["route"]]
        },
    })
    
print(routes_geojson)
```

```python
import json
import os

with open('../../../data/lsadata_hamburg.json') as f:
    lsas = json.load(f)
    
lsa_geojson = {
    "type": "FeatureCollection",
    "features": []
}
    
for lsa in lsas:
    geojson = lsa["thing"]["locations"][0]["location"]["geometry"]
    lsa_geojson["features"].append({
        "type": "Feature",
        "properties": {},
        "geometry": geojson,
    })
    
print(lsa_geojson)
```
